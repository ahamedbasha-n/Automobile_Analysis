# Exploratory Data Analysis of Automobile.

```python
import pandas as pd
import seaborn as sns
import matplotlib as plt
import warnings
warnings.filterwarnings('ignore')
sns.set(color_codes=True)
```

# Read the File:

```python
df=pd.read_csv("Automobile.csv") #df is data frame - variable
```

```python
df.head() 
```

```python
df.tail()
```

# Univariant Analysis

```python
sns.distplot(df['engine_size']) #with kernel density
plt.show()
```

```python
sns.distplot(df['engine_size'],kde=False,rug=True) #without kernel density
plt.show()
```

# Bi-Variant Analysis

```python
sns.jointplot(df['engine_size'],df['horsepower'])
plt.show()
```

```python
sns.jointplot(df['engine_size'],df['horsepower'],kind="hex")
plt.show()
```

```python
sns.jointplot(df['engine_size'],df['horsepower'],kind="reg")
plt.show()
```

```python
sns.jointplot(df['engine_size'],df['horsepower'],kind="kde")
plt.show()
```

```python
%matplotlib inline #prevents us to print the plot directly without using plt . show()
```

# Multi Variant

```python
sns.pairplot(df[['normalized_losses','engine_size','horsepower']])
```

# Analysis between a String and an Integer

```python
sns.stripplot(df['fuel_type'],df['horsepower'],jitter=True)
```

```python
sns.stripplot(df['fuel_type'],df['horsepower'],jitter=False)
```

```python
sns.swarmplot(df['fuel_type'],df['horsepower'])
```

```python
sns.boxplot(df['number_of_doors'],df['horsepower'])
```

```python
sns.boxplot(df['number_of_doors'],df['horsepower'],hue=df['fuel_type'])
```

```python
sns.barplot(df['body_style'],df['horsepower'])
```

```python
sns.barplot(df['body_style'],df['horsepower'],hue=df['engine_location'])
```

# Count Plot

```python
sns.countplot(df['body_style'])
```

```python
sns.pointplot(df['fuel_system'],df['horsepower'])
```

```python
sns.pointplot(df['fuel_system'],df['horsepower'],hue=df['number_of_doors'])
```

# Factor Plot

```python
sns.factorplot(x='fuel_type',y='horsepower',hue='number_of_doors',col='engine_location',data=df,kind='swarm')
```

#various types of kind input:{point,bar,count,box,violin,strip}

# Linear Regression

```python
sns.lmplot(x='horsepower',y='peak_rpm',data=df)
```
